####  Sound Generation Module  

Manages all audio notifications, ensuring timely alerts for critical events. Optimized for smooth execution and minimal disruption! 

````python
log_used_import("numpy")
log_used_import("scipy.io.wavfile")
log_used_import("scipy.signal.sawtooth")
log_used_import("IPython.display")
log_used_import("logging")
log_used_import("time")
log_used_import("random")

'''
--------------------------------------------------------------------------------
 MODULE INITIALIZATION: SOUND GENERATION MODULE
--------------------------------------------------------------------------------
 MODULE:        sound_generation
 STATUS:        HANDLING ALL AUDIO NOTIFICATIONS
 CHECKPOINT:    SOUND ALERT SYSTEM OPTIMIZED & READY
--------------------------------------------------------------------------------
'''

logging.info("[] Sound Generation Module Initialized!")

#  Global Sound Settings
SAMPLE_RATE = 44100  # Standard sample rate
VOLUME = 0.5  # Adjust volume for all sounds

# ==================================================================================
#  UNIVERSAL WAVE GENERATION FUNCTIONS
# ==================================================================================

def generate_waveform(frequency, duration, wave_type="sine"):
    """
    Generates an audio waveform of a given type (sine, square, triangle, sawtooth).
    """
    t = np.linspace(0, duration, int(SAMPLE_RATE * duration), endpoint=False)

    if wave_type == "sine":
        wave = np.sin(2 * np.pi * frequency * t)
    elif wave_type == "square":
        wave = np.sign(np.sin(2 * np.pi * frequency * t))
    elif wave_type == "triangle":
        wave = np.abs((2 * (t * frequency - np.floor(0.5 + t * frequency))) * 2 - 1)
    elif wave_type == "sawtooth":
        wave = sawtooth(2 * np.pi * frequency * t)
    else:
        raise ValueError(f"[] Invalid wave type: {wave_type}")

    return (wave * 32767 * VOLUME).astype(np.int16)


def save_wave(filename, wave):
    """
    Saves a waveform to a .wav file.
    """
    wav.write(filename, SAMPLE_RATE, wave)
    logging.info(f"[] Sound saved: {filename}")


def play_sound(filename):
    """
    Plays a saved .wav file.
    """
    ipd.display(ipd.Audio(filename, autoplay=True))

# ==================================================================================
#  ERROR ALERT SOUNDS
# ==================================================================================

def generate_error_sound():
    """
     Generates an error alert sound (harsh square wave beep).
    """
    wave = generate_waveform(frequency=440, duration=0.5, wave_type="square")
    save_wave("error_alert.wav", wave)

generate_error_sound()

def play_error(critical=False):
    """
     Plays the error alert sound.
     Standard errors: 1 beep
     Critical errors: 2 beeps
    """
    play_sound("error_alert.wav")
    if critical:
        time.sleep(0.6)
        play_sound("error_alert.wav")

# ==================================================================================
#  CHIME SOUND
# ==================================================================================

def generate_chime():
    """
     Generates a soft notification chime (sine wave).
    """
    wave = generate_waveform(frequency=800, duration=0.5, wave_type="sine")
    save_wave("chime.wav", wave)

generate_chime()

def play_chime():
    """Plays the chime sound."""
    play_sound("chime.wav")

# ==================================================================================
#  TRAINING & COMPLETION SOUNDS
# ==================================================================================

def generate_training_start():
    """
     Generates the training start sound (three ascending triangle pulses).
    """
    frequencies = [400, 500, 600]
    durations = [0.15, 0.15, 0.15]
    wave = np.concatenate([generate_waveform(f, d, "triangle") for f, d in zip(frequencies, durations)])
    save_wave("training_start.wav", wave)

generate_training_start()

def play_training_start():
    """Plays the training start sound."""
    play_sound("training_start.wav")


def generate_model_completion():
    """
     Generates the model completion sound (triangle arpeggio).
    """
    frequencies = [523.25, 659.25, 783.99]  # C5 - E5 - G5 (quick success tone)
    durations = [0.2, 0.2, 0.2]
    wave = np.concatenate([generate_waveform(f, d, "triangle") for f, d in zip(frequencies, durations)])
    save_wave("model_completion.wav", wave)

generate_model_completion()

def play_model_completion():
    """Plays the model completion sound."""
    play_sound("model_completion.wav")


def generate_training_complete():
    """
     Generates the final training loop completion sound (smooth chord pulse).
    """
    frequencies = [261.63, 329.63, 392.00]  # C4 - E4 - G4 (warm chord)
    duration = 0.4
    chord_wave = sum(generate_waveform(f, duration, "triangle") for f in frequencies)
    save_wave("training_complete.wav", chord_wave)

generate_training_complete()

def play_training_complete():
    """Plays the final training completion sound."""
    play_sound("training_complete.wav")

# ==================================================================================
#  WINNER FANFARE (Fanfare + Drumroll + Cymbal Crash)
# ==================================================================================

def generate_drumroll():
    """
     Generates a randomized snare drumroll effect.
    """
    duration = 1.2
    noise = np.random.uniform(-1, 1, int(SAMPLE_RATE * duration))
    fade_in = np.linspace(0, 1, len(noise))
    drumroll = noise * fade_in * 0.5
    return (drumroll * 32767).astype(np.int16)

def generate_fanfare():
    """
     Generates a triumphant fanfare using sawtooth waves.
    """
    notes = [523.25, 659.25, 783.99]  # C5, E5, G5
    duration = 0.3
    fanfare = np.concatenate([generate_waveform(note, duration, "sawtooth") for note in notes])
    return fanfare

def generate_cymbal():
    """
     Generates a cymbal crash (white noise burst).
    """
    duration = 0.8
    noise = np.random.uniform(-1, 1, int(SAMPLE_RATE * duration))
    fade_out = np.exp(-np.linspace(0, 5, len(noise)))
    cymbal = noise * fade_out
    return (cymbal * 32767).astype(np.int16)

def play_winner_fanfare():
    """
     Plays the full winner fanfare: Drumroll  Fanfare  Cymbal Crash.
    """
    logging.info("[] Playing Winner Fanfare... ")

    #  Generate and play drumroll
    save_wave("drumroll.wav", generate_drumroll())
    play_sound("drumroll.wav")
    time.sleep(1.2)

    #  Generate and play fanfare
    save_wave("fanfare.wav", generate_fanfare())
    play_sound("fanfare.wav")
    time.sleep(0.9)

    #  Generate and play cymbal crash
    save_wave("cymbal.wav", generate_cymbal())
    play_sound("cymbal.wav")

# ==================================================================================
#  TEST MODE: Plays Sounds for Debugging
# ==================================================================================

if DEBUG_MODE:
    logging.debug(" [DEBUG] Running test sound playback...")

    play_error()
    time.sleep(2)

    play_chime()
    time.sleep(2)

    play_training_start()
    time.sleep(2)

    play_model_completion()
    time.sleep(2)

    play_training_complete()
    time.sleep(2)

    play_winner_fanfare()
    time.sleep(2)
````

####  Error Handling System  

Manages system failures by logging errors and triggering alerts for smooth debugging.  Ensures stability and prevents disruptions during execution.

````python
log_used_import("os")
log_used_import("sys")
log_used_import("logging")
log_used_import("IPython.display")

'''
--------------------------------------------------------------------------------
 MODULE INITIALIZATION: ERROR HANDLING SYSTEM
--------------------------------------------------------------------------------
 MODULE:        error_handling
 STATUS:        MANAGING SYSTEM FAILURES & ALERTS
 CHECKPOINT:    ENSURING PROPER EXECUTION HANDLING
--------------------------------------------------------------------------------
'''

#  Error Testing Toggle (Enable only when testing)
DEBUG_TEST_ERRORS = False  #  Set to True to test errors, False to disable

def handle_error(error_message, critical=False):
    """
     Handles errors within the notebook.
     Standard errors log the issue and play a single error sound.
     Critical errors:
        - In **DEBUG MODE**, halts execution but does NOT disconnect runtime.
        - Outside **DEBUG MODE**, stops execution & disconnects GPU.
    """
    logging.error(f"[] ERROR: {error_message}")

    #  Play error sound (Critical errors beep twice)
    play_error(critical=critical)

    #  Critical Error Handling
    if critical:
        logging.critical("[] Critical error detected. Stopping execution...\n")

        if DEBUG_MODE:
            logging.warning("[] DEBUG MODE ENABLED: Execution halted, but runtime is still active.\n")
            raise RuntimeError("[] Critical error encountered. Execution halted, but runtime remains active.")

        else:
            logging.critical("[] Disconnecting GPU to save resources...\n")
            torch.cuda.empty_cache()
            tf.keras.backend.clear_session()


#  Conditional Error Testing (Runs only if enabled)
if DEBUG_MODE and DEBUG_TEST_ERRORS:
    try:
        raise ValueError("Test Standard Error")
    except Exception as e:
        handle_error(str(e))  # Standard error

    try:
        raise RuntimeError("Test Critical Error")
    except Exception as e:
        handle_error(str(e), critical=True)  # Critical error, stops execution
````

####  GPU-Based Batch Size Configuration  

Automatically detects **GPU availability** and adjusts the **batch size** for optimal training performance. This prevents memory overload and ensures efficient processing. 

````python
log_used_import("logging")
log_used_import("torch")
log_used_import("tensorflow")

'''
--------------------------------------------------------------------------------
 MODULE INITIALIZATION: GPU-BASED BATCH SIZE CONFIGURATION
--------------------------------------------------------------------------------
 MODULE:        gpu_batch_size_config
 STATUS:        AUTO-DETECTING GPU & ASSIGNING OPTIMAL BATCH SIZE
 CHECKPOINT:    ENSURING PERFORMANCE-OPTIMIZED TRAINING
--------------------------------------------------------------------------------
'''

def get_optimal_batch_size():
    """
     Determines the optimal batch size based on the detected GPU.
     Uses predefined batch size mappings for different GPU models.
     Supports PyTorch CUDA, Apple MPS, and CPU fallback.
    """
    try:
        if torch.cuda.is_available():
            gpu_name = torch.cuda.get_device_name(0).lower()

            #  Batch size mappings based on VRAM capacity
            batch_size_map = {
                "a100": 128,  # High-end GPU (large VRAM)
                "l4": 64,     # Balanced GPU
                "t4": 32,     # Low-end GPU
            }

            #  Assign batch size based on detected GPU
            for key, batch_size in batch_size_map.items():
                if key in gpu_name:
                    logging.info(f"[] GPU detected: {gpu_name} | Optimal batch size: {batch_size}")
                    return batch_size

            #  Default batch size for unrecognized GPUs
            logging.warning(f"[] Unrecognized GPU model: {gpu_name} | Using fallback batch size: 16")
            return 16

        elif torch.backends.mps.is_available():
            #  Apple Metal Performance Shaders (MPS) Support
            logging.info("[] Running on Apple MPS (Metal Performance Shaders) | Batch size: 32")
            return 32

        else:
            #  CPU Fallback
            logging.warning("[] No GPU detected! Running on CPU | Batch size: 16")
            return 16

    except Exception as e:
        handle_error(f"[] ERROR: GPU detection failed: {e} | Using safe batch size: 16")
        return 16  # Safe fallback batch size

#  Ensure TensorFlow does not over-allocate GPU memory
gpus = tf.config.experimental.list_physical_devices("GPU")
for gpu in gpus:
    tf.config.experimental.set_memory_growth(gpu, True)

#  Assign & Store Batch Size
BATCH_SIZE = get_optimal_batch_size()
GLOBAL_VARIABLES["BATCH_SIZE"] = BATCH_SIZE

logging.info(f"[] Final batch size assigned & stored in GLOBAL_VARIABLES: {BATCH_SIZE}")
````

####  Data Loader

Dynamically loads **training and testing data** as needed, optimizing memory usage and processing speed.  Ensures efficient data handling without unnecessary preloading. 

````python
log_used_import("logging")
log_used_import("tensorflow")
log_used_import("os")

'''
--------------------------------------------------------------------------------
 MODULE INITIALIZATION: DATA LOADER
--------------------------------------------------------------------------------
 MODULE:        load_dataset
 STATUS:        LOADING TRAINING, VALIDATION & TEST DATA
 CHECKPOINT:    VERIFYING & LOADING TF DATA PIPELINE
--------------------------------------------------------------------------------
'''

logging.info("[] Initializing Dataset Loader...\n")

def load_dataset(dataset_type="train"):
    """
     Loads dataset dynamically (Train/Validation/Test).
     Checks if dataset exists before attempting to load.
     Applies preprocessing & augmentation.
     Filters out hidden files like `.DS_Store`.
    """
    try:
        #  Ensure dataset paths are correctly defined
        dataset_dirs = {
            "train": GLOBAL_VARIABLES.get("TRAIN_DIR"),
            "validation": GLOBAL_VARIABLES.get("VAL_DIR"),
            "test": GLOBAL_VARIABLES.get("TEST_DIR")
        }

        dataset_dir = dataset_dirs.get(dataset_type)

        logging.info(f"[] Attempting to load dataset: {dataset_type.upper()} from {dataset_dir}")

        #  Check if directory exists
        if not dataset_dir or not os.path.exists(dataset_dir):
            logging.warning(f"[] WARNING: Dataset directory missing: {dataset_dir}")
            return None

        #  Remove any `.DS_Store` files
        for root, _, files in os.walk(dataset_dir):
            for file in files:
                if file == ".DS_Store":
                    os.remove(os.path.join(root, file))
                    logging.warning(f"[] Removed unexpected .DS_Store file from {root}")

        #  Check if dataset contains valid image subfolders
        subfolders = [folder for folder in os.listdir(dataset_dir) if not folder.startswith('.')]
        if not subfolders:
            logging.warning(f"[] WARNING: Dataset folder is empty after filtering: {dataset_dir}")
            return None

        logging.info(f"[] {dataset_type.upper()} Dataset Found! Subfolders: {subfolders}")

        #  Define Image Data Generators
        datagen_args = {"rescale": 1.0 / 255.0}
        if dataset_type == "train":
            datagen_args.update({
                "rotation_range": 15,
                "width_shift_range": 0.1,
                "height_shift_range": 0.1,
                "horizontal_flip": True
            })

        datagen = tf.keras.preprocessing.image.ImageDataGenerator(**datagen_args)

        #  Load dataset
        data_gen = datagen.flow_from_directory(
            dataset_dir,
            target_size=(128, 128),
            color_mode="rgb",
            batch_size=32,
            class_mode="categorical",
            shuffle=True
        )

        logging.info(f"[] Successfully loaded {dataset_type.upper()} dataset!")
        return data_gen

    except Exception as e:
        handle_error(f"[ ] ERROR LOADING {dataset_type.upper()} DATASET: {e}", critical=True)
        return None

import logging
import cv2
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator

def get_training_data():
    """
     Loads training & validation datasets dynamically with preprocessing.
    """
    logging.info("[] Calling load_dataset() for training & validation data...")

    train_data = load_dataset(dataset_type="train")
    val_data = load_dataset(dataset_type="validation")

    #  Check if dataset loaded properly
    if train_data is None:
        logging.warning("[] Training dataset failed to load! Check dataset paths.")
        return None, None
    if val_data is None:
        logging.warning("[] Validation dataset failed to load! Ensure correct preprocessing.")
        return None, None

    #  Image Preprocessing Function
    def preprocess_image(img):
        """
        Applies preprocessing steps:
        - Converts to grayscale (optional).
        - Normalizes pixel values.
        - Applies contrast normalization.
        """
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Convert to grayscale (optional)
        img = cv2.equalizeHist(img)  # Contrast normalization
        img = img / 255.0  # Normalize pixel values
        return img

    #  Data Augmentation for Training Set
    train_datagen = ImageDataGenerator(
        preprocessing_function=preprocess_image,  #  Apply preprocessing
        rotation_range=20,  #  Rotate up to 20 degrees
        horizontal_flip=True,  #  Flip horizontally
        width_shift_range=0.1,  #  Slight width shift
        height_shift_range=0.1,  #  Slight height shift
        shear_range=0.2,  #  Shearing
        zoom_range=0.2,  #  Random zoom
    )

    #  Validation Set: Only Normalization (No Augmentation)
    validation_datagen = ImageDataGenerator(preprocessing_function=preprocess_image)

    #  Apply Augmentation to the Datasets
    train_data = train_datagen.flow_from_directory(
        directory=GLOBAL_VARIABLES["TRAIN_DIR"],
        target_size=(128, 128),  # Match model input shape
        batch_size=GLOBAL_VARIABLES.get("BATCH_SIZE", 32),
        class_mode="categorical"
    )

    val_data = validation_datagen.flow_from_directory(
        directory=GLOBAL_VARIABLES["VAL_DIR"],
        target_size=(128, 128),
        batch_size=GLOBAL_VARIABLES.get("BATCH_SIZE", 32),
        class_mode="categorical"
    )

    return train_data, val_data  #  Now returns properly augmented datasets


def get_test_data():
    """
     Loads test dataset dynamically.
    """
    logging.info("[] Calling load_dataset() for test data...")
    test_data = load_dataset(dataset_type="test")

    if test_data is None:
        logging.warning("[] Test dataset failed to load! Verify dataset integrity.")

    return test_data
````

####  Test Evaluation  

Executes test runs on the registered model and logs **key performance metrics**.  Stores **predictions and accuracy scores** for final analysis. 

````python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import smart_resize
from sklearn.metrics import classification_report, confusion_matrix
import logging
import random

'''
--------------------------------------------------------------------------------
 MODULE INITIALIZATION: MODEL TESTING & EVALUATION
--------------------------------------------------------------------------------
 MODULE:        model_testing
 STATUS:        TESTING MODEL ON UNSEEN DATA
 CHECKPOINT:    EVALUATING MODEL PERFORMANCE ON TEST SET
--------------------------------------------------------------------------------
'''

logging.info("[] Running Model Evaluation on Test Data...")

#  Set the seed for reproducibility
SEED_VALUE = 42
random.seed(SEED_VALUE)
np.random.seed(SEED_VALUE)
tf.random.set_seed(SEED_VALUE)
logging.info(f"[] Random Seed Applied for Consistent Results! (Seed Value: {SEED_VALUE})")

#  Define Correct Class Names
CLASS_NAMES = ["Happy", "Sad", "Neutral", "Surprise"]

try:
    #  Ensure TEST_RESULTS exist
    if "TEST_RESULTS" not in GLOBAL_VARIABLES:
        GLOBAL_VARIABLES["TEST_RESULTS"] = {}

    #  Load test dataset dynamically
    test_data = get_test_data()

    if test_data is None:
        raise ValueError("[] Test dataset is missing! Ensure test data is properly loaded.")

    logging.info(f"[] Test dataset type: {type(test_data)}")  #  Debugging print

    #  Extract images and labels from DirectoryIterator properly
    X_test, y_true = [], []

    #  Iterate through the test dataset to retrieve all batches
    for i in range(len(test_data)):
        batch_images, batch_labels = test_data[i]  #  Retrieve batch properly
        X_test.append(batch_images)
        y_true.append(batch_labels)

    #  Convert to numpy arrays
    X_test = np.vstack(X_test)  #  Stack all image batches
    y_true = np.vstack(y_true)  #  Stack all label batches
    y_true = np.argmax(y_true, axis=1)  #  Convert one-hot encoding to class labels

    logging.info(f"[] Extracted {X_test.shape[0]} test images and labels.")

    #  Ensure model exists in GLOBAL_VARIABLES or use global variable as fallback
    model = GLOBAL_VARIABLES.get("MODEL", None) or globals().get("vgg16_model", None)

    if model is None:
        raise ValueError("[] Model is missing! Ensure the model has been trained and stored in GLOBAL_VARIABLES.")

    #  Retrieve expected input shape from model
    expected_input_shape = model.input_shape[1:4]  # Extracts (height, width, channels)
    logging.info(f"[] Model expects input shape: {expected_input_shape}")

    #  Check if test images need resizing
    if X_test.shape[1:4] != expected_input_shape:
        logging.warning(f"[] Test images are {X_test.shape[1:4]} but expected {expected_input_shape}. Resizing now...")
        X_test = np.array([smart_resize(img, expected_input_shape[:2]) for img in X_test])  #  Resize properly
        logging.info(f"[] Resized all test images to {expected_input_shape}")

    #  Generate predictions on test set
    logging.info("[] Generating predictions on test dataset...")
    y_pred_probs = model.predict(X_test)  #  Get softmax probabilities
    y_pred = np.argmax(y_pred_probs, axis=1)  #  Convert to class labels

    #  Store predictions & ground truth labels
    GLOBAL_VARIABLES["TEST_RESULTS"]["y_pred"] = y_pred
    GLOBAL_VARIABLES["TEST_RESULTS"]["y_true"] = y_true

    #  Compute Confusion Matrix & Classification Report
    cm = confusion_matrix(y_true, y_pred)
    GLOBAL_VARIABLES["TEST_RESULTS"]["confusion_matrix"] = cm

    class_report_dict = classification_report(y_true, y_pred, target_names=CLASS_NAMES, output_dict=True)
    GLOBAL_VARIABLES["TEST_RESULTS"]["classification_report"] = class_report_dict
    logging.info("[] Classification Report Successfully Generated!")

    #  Extract Precision, Recall, F1-score
    precisions = [class_report_dict[name]["precision"] for name in CLASS_NAMES]
    recalls = [class_report_dict[name]["recall"] for name in CLASS_NAMES]
    f1_scores = [class_report_dict[name]["f1-score"] for name in CLASS_NAMES]

    #  Retrieve Validation Accuracy & Store it for Future Use
    if "val_accuracy" not in GLOBAL_VARIABLES:
        val_accuracy = TRAINING_HISTORY.history["val_accuracy"][-1] if "val_accuracy" in TRAINING_HISTORY.history else None
        GLOBAL_VARIABLES["val_accuracy"] = val_accuracy
    else:
        val_accuracy = GLOBAL_VARIABLES["val_accuracy"]

    #  Log Test Metrics
    test_accuracy_str = f"{np.mean(y_pred == y_true) * 100:.2f}%"  #  Compute accuracy dynamically
    precision_str = f"{np.mean(precisions) * 100:.2f}%" if precisions else "N/A"
    recall_str = f"{np.mean(recalls) * 100:.2f}%" if recalls else "N/A"
    f1_score_str = f"{np.mean(f1_scores) * 100:.2f}%" if f1_scores else "N/A"
    val_accuracy_str = f"{val_accuracy * 100:.2f}%" if isinstance(val_accuracy, float) else "N/A"

    logging.info(f"""
    --------------------------------------------------------------------------------
     MODEL TEST RESULTS
    --------------------------------------------------------------------------------
     TEST ACCURACY:      {test_accuracy_str}
     PRECISION:          {precision_str}
     RECALL:             {recall_str}
     F1 SCORE:           {f1_score_str}
     VALIDATION ACCURACY: {val_accuracy_str}
    --------------------------------------------------------------------------------
    """)

    logging.info("[] Model Testing Completed Successfully!")

except Exception as e:
    handle_error(f"[ ] ERROR DURING MODEL TESTING: {e}", critical=True)
````

````python
'''
--------------------------------------------------------------------------------
 MODULE INITIALIZATION: TRAINING RESULTS VISUALIZATION
--------------------------------------------------------------------------------
 MODULE:        visualize_training_results
 STATUS:        LOADING & COMPARING TRAINING RESULTS
 CHECKPOINT:    GENERATING PERFORMANCE GRAPHS
--------------------------------------------------------------------------------
'''

logging.info("[] Generating Training Results Visualizations...")

#  Define Correct Directory
RESULTS_DIR = "/content/drive/MyDrive/Colab Notebooks/VGG16_Training_Results/"

#  Function to Convert Seconds to H:M:S
def convert_seconds_to_hms(seconds):
    """Converts seconds (float) to H:M:S format"""
    if seconds is None or seconds == 0:
        return "00:00:00"
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    secs = int(seconds % 60)
    return f"{hours:02d}:{minutes:02d}:{secs:02d}"

#  Ensure Directory Exists
if not os.path.exists(RESULTS_DIR):
    logging.error(f"[] Directory not found: {RESULTS_DIR}")
    raise FileNotFoundError(f"Results directory not found: {RESULTS_DIR}")

#  Load JSON Files for All Iterations
result_files = sorted([f for f in os.listdir(RESULTS_DIR) if f.startswith("iteration_") and f.endswith(".json")])
iterations = []
train_accuracies, val_accuracies, test_accuracies = [], [], []
train_losses, test_losses, training_times = [], [], []
model_scores_list = []

logging.info(f"[] Found {len(result_files)} JSON files for iterations.")

#  Extract Data from Each Iteration
for file in result_files:
    iteration_path = os.path.join(RESULTS_DIR, file)

    try:
        with open(iteration_path, "r") as json_file:
            data = json.load(json_file)

        iteration_number = data.get("iteration", None)
        test_results = data.get("test_results", {})

        if iteration_number is None or not test_results:
            logging.warning(f"[] Skipping {file} due to missing data!")
            continue

        iterations.append(iteration_number)
        train_accuracies.append(test_results.get("Train Accuracy", np.nan))
        val_accuracies.append(test_results.get("Validation Accuracy", np.nan))  #  Now properly retrieving it!
        test_accuracies.append(test_results.get("Test Accuracy", np.nan))
        train_losses.append(test_results.get("Train Loss", np.nan))
        test_losses.append(test_results.get("Test Loss", np.nan))

        #  Extract and Convert Training Time
        raw_training_time = test_results.get("Training Time (s)", 1)
        try:
            raw_training_time = float(raw_training_time)
        except ValueError:
            logging.warning(f"[] Invalid training time in {file}, defaulting to 1 second.")
            raw_training_time = 1

        training_times.append(raw_training_time)

    except Exception as e:
        logging.error(f"[] Failed to process {file}: {e}")
        continue

#  Ensure Data Integrity (Convert to NumPy Arrays)
train_accuracies = np.array(train_accuracies, dtype=np.float64)
val_accuracies = np.array(val_accuracies, dtype=np.float64)
test_accuracies = np.array(test_accuracies, dtype=np.float64)
train_losses = np.array(train_losses, dtype=np.float64)
test_losses = np.array(test_losses, dtype=np.float64)
training_times = np.array(training_times, dtype=np.float64)

#  Calculate Total Score for Each Model Iteration
for i in range(len(iterations)):
    best_count = 0
    if train_accuracies[i] == np.nanmax(train_accuracies): best_count += 1
    if val_accuracies[i] == np.nanmax(val_accuracies): best_count += 1  #  Now considered!
    if test_accuracies[i] == np.nanmax(test_accuracies): best_count += 1
    if train_losses[i] == np.nanmin(train_losses): best_count += 1  # Lower loss is better
    if test_losses[i] == np.nanmin(test_losses): best_count += 1  # Lower loss is better
    model_scores_list.append(best_count)

#  Determine Best Model Iteration
if model_scores_list:
    best_iteration_index = np.argmax(model_scores_list)  # Find best model index
    best_iteration = iterations[best_iteration_index]  # Retrieve best iteration number
    best_model_score = model_scores_list[best_iteration_index]  # Best model's total score
    best_training_time = training_times[best_iteration_index]
else:
    logging.error("[] No valid model scores were found.")
    best_iteration, best_model_score, best_training_time = None, None, None

#  Debugging Output
logging.info(f"[] Best iteration determined: Iteration {best_iteration} with score {best_model_score}")
logging.info(f"[DEBUG] Best Model Training Time: {best_training_time} seconds  {convert_seconds_to_hms(best_training_time)}")

#  Generate Visualizations
fig, axes = plt.subplots(4, 1, figsize=(12, 20))

#  Apply Global Colormap
COLORMAP = "nipy_spectral"
cmap = plt.get_cmap(COLORMAP, len(iterations))

#  1. Training, Validation & Test Accuracy Over Iterations
axes[0].plot(iterations, train_accuracies, marker="o", linestyle="-", label="Train Accuracy", color="blue", alpha=0.6)
axes[0].plot(iterations, val_accuracies, marker="s", linestyle="--", label="Validation Accuracy", color="orange", alpha=0.6)
axes[0].plot(iterations, test_accuracies, marker="^", linestyle=":", label="Test Accuracy", color="green", alpha=0.6)
axes[0].set_title("Training, Validation & Test Accuracy Across Iterations")
axes[0].set_xlabel("Iteration")
axes[0].set_ylabel("Accuracy (%)")
axes[0].set_xticks(iterations)
axes[0].legend()
axes[0].grid(True, linestyle="--", alpha=0.6)

#  2. Training & Test Loss Over Iterations
axes[1].plot(iterations, train_losses, marker="o", linestyle="-", label="Train Loss", color="red", alpha=0.6)
axes[1].plot(iterations, test_losses, marker="s", linestyle="--", label="Test Loss", color="purple", alpha=0.6)
axes[1].set_title("Training & Test Loss Across Iterations")
axes[1].set_xlabel("Iteration")
axes[1].set_ylabel("Loss")
axes[1].set_xticks(iterations)
axes[1].legend()
axes[1].grid(True, linestyle="--", alpha=0.6)

#  3. Total Score
bar_colors = [cmap(i / len(iterations)) for i in range(len(iterations))]
axes[2].bar(iterations, model_scores_list, color=bar_colors, edgecolor="black")
axes[2].set_title("Overall Model Score")
axes[2].set_xlabel("Iteration")
axes[2].set_ylabel("Total Score")
axes[2].set_xticks(iterations)
axes[2].grid(True, linestyle="--", alpha=0.6)

#  4. Training Time
axes[3].bar(iterations, training_times, color=bar_colors, edgecolor="black", label="Training Time (h:m:s)", alpha=0.6)
axes[3].set_title("Training Time Across Iterations")
axes[3].set_xlabel("Iteration")
axes[3].set_ylabel("Time (Seconds)")
axes[3].set_xticks(iterations)
axes[3].grid(True, linestyle="--", alpha=0.6)

#  Adjust Layout and Show Plots
plt.tight_layout()
plt.show()
````

####  Display Model Iterations  

Loads and formats model summaries  for easy comparison across iterations. Helps visualize architectural changes and improvements over time.

````python
log_used_import("os")
log_used_import("json")
log_used_import("logging")
log_used_import("IPython.display")

'''
--------------------------------------------------------------------------------
 MODULE INITIALIZATION: DISPLAY MODEL ITERATIONS
--------------------------------------------------------------------------------
 MODULE:        display_iterations
 STATUS:        LOADING & FORMATTING MODEL SUMMARIES
 CHECKPOINT:    ENSURING READABLE OUTPUT FOR COMPARISON
--------------------------------------------------------------------------------
'''

logging.info("[] Loading all model iterations...")

#  Define storage directory
ITERATION_DIR = "/content/drive/MyDrive/Colab Notebooks/VGG16_Training_Results/"
os.makedirs(ITERATION_DIR, exist_ok=True)  #  Ensure directory exists

#  Retrieve all iterations sorted numerically
iteration_files = sorted(
    [f for f in os.listdir(ITERATION_DIR) if f.startswith("iteration_") and f.endswith(".json")],
    key=lambda x: int(x.split("_")[1].split(".")[0])
)

if not iteration_files:
    logging.warning("[] No model iterations found! Ensure at least one exists before running this cell.")
    print(" No model iterations found! Please check your saved files.")
else:
    logging.info(f"[] Found {len(iteration_files)} iterations. Displaying results...\n")

    for iteration_file in iteration_files:
        iteration_path = os.path.join(ITERATION_DIR, iteration_file)

        try:
            with open(iteration_path, "r") as f:
                iteration_data = json.load(f)

            #  Extract Data Safely
            iteration_number = iteration_data.get("iteration", "Unknown")
            model_summary = iteration_data.get("model_metadata", {}).get("summary", "No summary available")
            model_hash = iteration_data.get("model_metadata", {}).get("hash", "No hash available")
            training_history = iteration_data.get("training_history", {})
            test_results = iteration_data.get("test_results", {})

            #  Display Iteration Header
            print("\n" + "=" * 80)
            print(f" MODEL ITERATION {iteration_number}")
            print("=" * 80)

            #  Display Model Hash
            print("\n MODEL HASH:")
            print("-" * 80)
            print(model_hash)
            print("-" * 80)

            #  Display Model Summary
            print("\n MODEL SUMMARY:")
            print("-" * 80)
            print(model_summary)
            print("-" * 80)

            #  Display Training History (if available)
            if training_history:
                print("\n TRAINING HISTORY:")
                print("-" * 80)
                for metric, values in training_history.items():
                    print(f" {metric}: {values[-1] if values else 'N/A'}")  #  Show final value of each metric
                print("-" * 80)
            else:
                print("\n TRAINING HISTORY: [] No history recorded.")

            #  Display Test Results (if available)
            if test_results:
                print("\n TEST RESULTS:")
                print("-" * 80)
                print(f" Test Accuracy: {test_results.get('test_accuracy', 'N/A')}")
                print(f" Test Loss: {test_results.get('test_loss', 'N/A')}")
                print(f" Execution Time: {test_results.get('training_time', 'N/A')}")
                print("-" * 80)
            else:
                print("\n TEST RESULTS: [] No test results recorded.")

            #  Add a visual break between iterations
            ipd.display(ipd.HTML("<br><hr style='border: 1px solid gray;'><br>"))

        except Exception as e:
            logging.error(f"[] ERROR: Failed to load iteration {iteration_file}! {e}")
            print(f" ERROR: Could not read iteration file {iteration_file}. Check the file integrity.\n")
````

####  Import Validation Check  

Scans all imported libraries to detect unused or redundant imports. Helps optimize performance by keeping the environment clean and efficient.


````python
log_used_import("importlib")
log_used_import("sys")

'''
--------------------------------------------------------------------------------
 MODULE FINALIZATION: IMPORT VALIDATION CHECK
--------------------------------------------------------------------------------
 MODULE:        import_validation
 STATUS:        CROSS-CHECKING IMPORT USAGE
 CHECKPOINT:    IDENTIFYING UNUSED IMPORTS
--------------------------------------------------------------------------------
'''

def validate_imports():
    """
     Compares `IMPORT_TRACKER` against `USED_IMPORTS` to detect unnecessary imports.
     Uses both manual tracking (`log_used_import`) and `importlib` for validation.
     Ensures all imports are being utilized in execution.
    """
    global USED_IMPORTS

    #  Automatically detect referenced modules in the global scope
    detected_imports = {name.split(".")[0] for name, obj in globals().items() if isinstance(obj, type(sys)) or callable(obj)}

    #  Combine auto-detected imports with manually logged ones
    USED_IMPORTS |= detected_imports

    #  Identify unused imports
    unused_imports = IMPORT_TRACKER - USED_IMPORTS

    #  Display results
    if unused_imports:
        logging.warning(f"[] WARNING: The following imports are unused and may be removed:\n{unused_imports}")
    else:
        logging.info("[] All imported modules are in use.")

#  Run final validation check
validate_imports()
````
